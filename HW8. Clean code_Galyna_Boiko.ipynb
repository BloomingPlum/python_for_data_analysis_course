Цей варіант компактніший, використовуються більш зрозумілі назви змінних, замість двох циклів тільки один у вигляді list comprehension і він дає правильний результат. 

````python
def calculate_variance(numbers):
    if not numbers:  # Check for empty list
        return 0
    
    sum_list = sum(numbers)
    mean = sum_list / len(numbers)
    
    sum_squares = sum(num**2 for num in numbers)
    mean_squares = sum_squares / len(numbers)
    
    return mean_squares - mean**2
````

````python
print (calculate_variance(input_list)) 
````
